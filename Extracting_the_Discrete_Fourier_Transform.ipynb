```python
violin_c4,sr=librosa.load(os.path.join(violin_sound_file))
sax_c4,_=librosa.load(os.path.join(sax_sound_file))
piano_c5,_=librosa.load(os.path.join(piano_sound_file))
noise_c4,_=librosa.load(os.path.join(noise_sound_file))
```

```python
# default sampling rate
sr
```

```python
violin_c4.shape
```

```python
#module contains ffft and various implementations in it and next is the classical fft
violin_ft=np.fft.fft(violin_c4)
sax_ft=np.fft.fft(sax_c4)
piano_ft=np.fft.fft(piano_c5)
noise_ft=np.fft.fft(noise_c4)
```

```python
violin_ft.shape
```

```python
violin_ft[0]
# complex number-real and img part-magnitude and phase
```

```python
magnitude_spectrum_violin=np.abs(violin_ft)
magnitude_spectrum_sax=np.abs(sax_ft)
magnitude_spectrum_piano=np.abs(piano_ft)
magnitude_spectrum_noise=np.abs(noise_ft)
```

```python
magnitude_spectrum_violin[0]
# Only magnitude associated with initial frequency
```

```python
import matplotlib.pyplot as plt
```

```python
def plot_magnitude_spectrum(signal,title,sr,f_ratio=0.1):
  ft=np.fft.fft(signal)
  magnitude_spectrum=np.abs(ft)


# plot amplitude spectrum
f_ratio=0.1
plt.figure(figsize=(10,5))
frequencies=np.linspace(0,sr,len(magnitude_spectrum_violin))
num_frequency_bins=int(len(frequencies)*f_ratio)
plt.plot(frequencies[:num_frequency_bins],magnitude_spectrum_violin[:num_frequency_bins])
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
plt.title("Frequency vs magnitude")
plt.show()
# first-fundamental frequency ,next are harmonics
```
